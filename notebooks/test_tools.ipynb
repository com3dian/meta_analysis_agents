```python
from dataclasses import dataclass
from typing import Any

from src.tools.context_tools import register_context
from src.tools.xml_tagging import XMLTagFromFieldValuesTool

# Minimal dummy context that matches the interface used by the tool
@dataclass
class DummyContext:
    content: str

    def read_resource(self, resource: str, *args, **kwargs) -> Any:
        # Ignore resource name for this simple test
        return self.content

dummy_doc = """
This study was conducted at Tongshan experimental station, Xuzhou City, Jiangsu Province, China
during the 2020–2021 growing season. Winter wheat was planted, and the mean yield was 529.58 g/m2.
"""

# Register context so the tool can look it up
ctx = DummyContext(content=dummy_doc)
register_context("test_ctx", ctx)

tool = XMLTagFromFieldValuesTool()

result = tool.invoke({
    "context_key": "test_ctx",
    "resource": "dummy_resource",
    "field_value_pairs": [
        ("location", "Tongshan experimental station, Xuzhou City, Jiangsu Province, China"),
        ("year", "2020–2021"),
        ("crop_type", "Winter wheat"),
        ("yield_value", "529.58 g/m2"),
    ],
})

print(result)
```
